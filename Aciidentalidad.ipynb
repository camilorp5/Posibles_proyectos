```python
import re

def convertir_a_24h(hora):
    hora = hora.strip().replace(" ", "").replace(".", "").upper()
    formato_12h_regex = r"^\d{1,2}:\d{2}:\d{2}(AM|PM)$"

    if re.match(formato_12h_regex, hora):  # Si el formato es válido
        return pd.to_datetime(hora, format="%I:%M:%S%p").strftime("%H:%M:%S")
    else:
        return hora
df["Hora_24h"] = df["HORA"].apply(convertir_a_24h)

df
```

```python
def cambiar_hora(x):
  if "m" in str(x):
    return str(x)
  elif int(str(x)[:2])-12>0 :
    return f'{int(str(x)[:2])-12}{str(x)[2:8]} p.m.'
  else:
    return f'{int(str(x)[:2])}{str(x)[2:8]} a.m.'
```

```python
df_prueba['hora'].apply(cambiar_hora)
```
